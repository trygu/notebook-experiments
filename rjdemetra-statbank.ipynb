## Demo av RJDemetra + PxWebApiData

```r
library(PxWebApiData)
library(RJDemetra)
```

Henter ujustert volumindeks fra NACE 47.1 - Butikkhandel med bredt vareutvalg, fra Statistikkbanken (https://www.ssb.no/statbank/table/07129/). Alle m√•neder fra 2000 til i dag. 

```r
innData <- ApiData(7129, NACE = "47.1", ContentsCode = "VolumUjustert", Tid = TRUE)
ujustert <- ts(innData$dataset$value, freq = 12, start = c(2000, 1))
```

```r
x13_model <- x13(ujustert)
```

```r
plot(x13_model, type_chart = c("sa-trend", "cal-seas-irr"))
plot(x13_model$decomposition)
plot(x13_model$regarima)
```
